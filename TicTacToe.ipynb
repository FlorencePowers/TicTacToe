```python
def print_the_board(gameBoard):
    
    row = ['   |   |   ', '-----------']

    rowOne   = gameBoard['row1']
    rowTwo   = gameBoard['row2']
    rowThree = gameBoard['row3']

    print(' ')
    print(f' {rowOne[0]} | {rowOne[1]} | {rowOne[2]} ')
    print(row[1])
    print(f' {rowTwo[0]} | {rowTwo[1]} | {rowTwo[2]} ')
    print(row[1])
    print(f' {rowThree[0]} | {rowThree[1]} | {rowThree[2]} ')   
```

```python
def board_grid():
    
    rowOne   = [' ', ' ', ' ']
    rowTwo   = [' ', ' ', ' ']
    rowThree = [' ', ' ', ' ']
    
    grid = {"row1": rowOne, "row2": rowTwo, "row3":rowThree}
    
    return grid
```

```python
def playGame(gameBoard, player):
    
    nextPlayerGoes = False
    
    choice = players_choice(player)

    #keep playing until valid choice made
    needToKeepPlaying = True
    
    while needToKeepPlaying:
        #check if option selected is available
        if canPlay(gameBoard, player, choice):
            clear_output()
            gameBoard = updatePlay(gameBoard, player, choice)
            print_the_board(gameBoard)
            if isAWinner(gameBoard, player) or isBoardFull(gameBoard):
                nextPlayerGoes = False
                needToKeepPlaying = False 
            else:
                nextPlayerGoes = True
                needToKeepPlaying = False
        else:
            print("Choice was already taken, try another number from 1-9")
            choice = players_choice(player)
            
    return (nextPlayerGoes, gameBoard)
```

```python
def isBoardFull(board):

    row1 = board.get('row1')
    row2 = board.get('row2')
    row3 = board.get('row3')
    
    if ((row1[0] == ' ' or row1[1] == ' ' or row1[2] == ' ') or
        (row2[0] == ' ' or row2[1] == ' ' or row2[2] == ' ') or
        (row3[0] == ' ' or row3[1] == ' ' or row3[2] == ' ')):
        return False
    else:
        return True
    
```

```python
def isAWinner(board, player):
    
    winner = False
    
    row1 = board.get('row1')
    row2 = board.get('row2')
    row3 = board.get('row3')
    
    if ((row1[0] == player and row1[1] == player and row1[2] == player) or
        (row2[0] == player and row2[1] == player and row2[2] == player) or
        (row3[0] == player and row3[1] == player and row3[2] == player)):       
        winner = True
    elif ((row1[0]  == player and row2[0]  == player and row3[0] == player) or
          (row1[1]  == player and row2[1]  == player and row3[1] == player) or
          (row1[2]  == player and row2[2]  == player and row3[2] == player)):
        winner = True
    elif ((row1[0]  == player and row2[1]  == player and row3[2] == player) or
          (row1[2]  == player and row2[1]  == player and row3[0]  == player)):
        winner = True
        
    return winner    
```

```python
def players_choice(player):
    
    playersChoice = ' '
    goodAnswer = False
    validAnswers = range(1,10)

    while playersChoice.isdigit() == False and goodAnswer == False:

        playersChoice = input(f"\nYOUR TURN PLAYER {player}: Enter a number between 1-9: ")
        
        if playersChoice.isdigit() and int(playersChoice) in validAnswers:
            goodAnswer = True
        else:
            playersChoice = input("Invalid response, enter a number between 1-9: ")
                  
    return int(playersChoice)
            
```

```python
def canPlay(gameBoard, player, choice):
    
    key = ''
    theIndex = 0

    if choice in [1,2,3]:
        key = 'row1'
    elif choice in [4,5,6]:
        key = 'row2'
    else:
        key = 'row3'
    
    values = gameBoard.get(key)
    
    if choice in [1, 4, 7]:
        theIndex = 0
    elif choice in [2, 5, 8]:
        theIndex = 1
    else:
        theIndex = 2

    if values[theIndex] not in ['O', 'X']:
        return True
    else:
        return False    
```

```python
def updatePlay(gameBoard, player, choice):
    key = ''
    theIndex = 0

    if choice in [1,2,3]:
        key = 'row1'
    elif choice in [4,5,6]:
        key = 'row2'
    else:
        key = 'row3'
  
    if choice in [1, 4, 7]:
        theIndex = 0
    elif choice in [2, 5, 8]:
        theIndex = 1
    else:
        theIndex = 2
        
    values = gameBoard[key]
    values[theIndex] = player
    gameBoard[key]=values
    
    return gameBoard
```
